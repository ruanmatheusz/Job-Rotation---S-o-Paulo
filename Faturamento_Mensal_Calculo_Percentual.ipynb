```python
import pandas as pd

dados = [('SP', 67.83643),
('RJ' , 36.67866),
('MG' , 29.22988),
('ES' , 27.16548),
('Outros' , 19.84953)]

df = pd.DataFrame(dados, columns = ['estado', 'receita']).round(decimals = 3)
df['porcentagem'] = (df['receita'] / df['receita'].sum()) * 100
df
```
